# Scraping Images from Website

In this project. I was able to create a scraper for the website www.pngmart.com. This scraper will allow you to scrape 10 images at a time. I was able to learn more about requests and how they work with BeautifulSoup. Images inside of this project were the first downloaded when using the scraper to confirm it works. 

```python
import requests
from bs4 import BeautifulSoup
```

```python
site_map = 'http://www.pngmart.com/sitemapindex.xml'
```

```python
response = requests.get(site_map)
```

```python
xml = response.text
```

```python
soup = BeautifulSoup(xml, 'xml')
```

```python
# Looking up the loc on the site
soup.find_all('loc')
```

```python
site_maps = []
for loc in soup.find_all('loc'):
    url = loc.text
    if 'part' in url:
        print (loc.text)
        site_maps.append(url)
```
